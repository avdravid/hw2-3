```python
import numpy as np
import matplotlib.pyplot as plt

# Particle counts
particles = [1000, 10000, 50000, 100000, 500000, 1000000]

# Execution times for each method (in seconds)
gpu_times = [0.0417766, 0.0439388, 0.0582753, 0.0806736, 0.283327, 0.578849]
cpu_mpi_times = [0.0660043, 0.0720346, 0.109169, 0.137796, 0.484336, 1.00271]
cpu_openmp_times = [0.11985, 0.42687, 1.83314, 2.78, 9.48755, 18.1205]

# Convert times to ms or s where appropriate
def format_time(time):
    if time < 1e-6:
        return f"{time * 1e9:.2f} ns"
    elif time < 1e-3:
        return f"{time * 1e6:.2f} Âµs"
    elif time < 1:
        return f"{time * 1e3:.2f} ms"
    else:
        return f"{time:.2f} s"

# Plotting the data
plt.figure(figsize=(8, 6))
plt.plot(particles, gpu_times, marker='o', linestyle='-', label="GPU Optimized")
plt.plot(particles, cpu_mpi_times, marker='s', linestyle='-', label="CPU MPI (N=2, 64 tasks/node)")
plt.plot(particles, cpu_openmp_times, marker='^', linestyle='-', label="CPU OpenMP (64 threads)")

plt.yscale('log')
plt.xscale('log')
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

plt.xlabel("# Particles")
plt.ylabel("Time (s)")
plt.title("Performance Comparison: GPU vs. CPU MPI vs. CPU OpenMP")
plt.legend()

# Annotate with execution times
for i, n in enumerate(particles):
    plt.text(n, gpu_times[i], format_time(gpu_times[i]), fontsize=8, verticalalignment='bottom')
    plt.text(n, cpu_mpi_times[i], format_time(cpu_mpi_times[i]), fontsize=8, verticalalignment='bottom')
    plt.text(n, cpu_openmp_times[i], format_time(cpu_openmp_times[i]), fontsize=8, verticalalignment='bottom')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Particle counts and corresponding GPU Optimized execution times (in seconds)
particles = [1000, 5000, 10000, 25000, 50000, 75000, 100000, 1000000, 5000000, 10000000]
gpu_optimized_times = [0.0417766, 0.0434687, 0.0439388, 0.0487412, 0.0582753, 0.0685741, 0.0806736, 0.578849, 3.59703, 7.82693]

# Plot data
plt.figure(figsize=(8, 6))
plt.plot(particles, gpu_optimized_times, marker='o', linestyle='-', color='b', label="GPU Optimized")

# Log-log scale
plt.xscale('log')
plt.yscale('log')

# Labels and title
plt.xlabel("Number of Particles")
plt.ylabel("Execution Time (s)")
plt.title("GPU Optimized Performance")

# Annotate data points with execution times
for i, txt in enumerate(gpu_optimized_times):
    plt.annotate(f"{txt:.2f}s", (particles[i], gpu_optimized_times[i]), textcoords="offset points", xytext=(0, 5), ha='center')

plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.show()
```
